```python
import numpy as np
import numpyro
import numpyro.distributions as dist
import matplotlib.pyplot as plt
from jax import random
from jax.nn import softplus

# Set a random seed for reproducibility
numpyro.set_host_device_count(1)  # Ensure single-threaded execution
np.random.seed(42)

# Generate an RNG key
rng_key = random.PRNGKey(0)

# Define a function to sample noise
def sample_noise(rng_key):
    return dist.Normal(0, 1).sample(rng_key, sample_shape=(1000,))

# Sample noise from a normal distribution
x10_noise = sample_noise(rng_key)

# Apply softplus transformation
x10_noise_softplus = softplus(x10_noise)


# Plot the distribution of the noise before and after softplus
plt.figure(figsize=(12, 6))

# Before softplus
plt.subplot(1, 2, 1)
plt.hist(x10_noise, bins=50, density=True, alpha=0.6, color="blue", label="Before Softplus")
plt.title("Distribution of Noise (Before Softplus)")
plt.xlabel("Noise")
plt.ylabel("Density")
plt.legend()

# After softplus
plt.subplot(1, 2, 2)
plt.hist(x10_noise_softplus, bins=50, density=True, alpha=0.6, color="red", label="After Softplus")
plt.title("Distribution of Noise (After Softplus)")
plt.xlabel("Noise")
plt.ylabel("Density")
plt.legend()

plt.tight_layout()
plt.show()
```
